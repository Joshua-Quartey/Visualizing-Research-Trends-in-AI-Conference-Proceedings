## Introduction

**Conference proceedings**\
Conferences are a major mechanism in the professional lives of  researchers facilitating the dissemination of research findings and the advancement of the field through collaboration. 

Information is typically delivered through presentations, posters, workshops  or panel discussions.
Work presented gets published in research literature in the form of conference proceedings or journal articles.

Views on how conference proceedings are weighed as evidence of scholarly activity vary by discipline considering they are generally subjected to different review standards compared to journals, are structured differently,  and cited less frequently.

Conference publications in computer science, however seem to have high scholarly impact, unlike other disciplines where journals are dominantly used for communicating scientific findings.

## Motivation

Analysing the contribution of conference papers to the scientific literature is not that well documented and is the  aim of this project.

Insights from this analysis can help 
explore the impact of proceedings papers on future research as well as help researchers identify potential collaborators and experts who are working on complementary areas of research and presenting new research at scholarly events.

This project aims to measure the interest in topics to help document how publications are being used by the academic community

The insights of this project might also be useful to policy makers responsible for recruitment, academic promotion, or research grant awards by providing them some form of quantitative assessment of their publication output and citation impact when taking decisions

## Datasets 

The dataset was obtained from Elsevier's Scopus, a multidisciplinary scientifc citation index. The data consists of publications tagged as "proceeding papers" and Artificial Intelligence related research topics in Computer Science from 2016 to 2019.

## Analysis

**Methodology**\
The method followed for analyis was as follows:
First, publications were extracted from Scopus  and organised by country. The second step was to analyse the patterns from collaboration between countries and to track publishing behaviour

Data was accessed using the University of Calgary's Library subscription to Scopus. Data extraction was limited to at most 20,000 records per search and so the publications retrieved for each year in the 2016 to 2024 period were of equal length. Scopus data were extracted in February 2025.

The papers used in this study were those with document types proceedings paper and proceeding from Scopus.

The analysis excluded editorials, and other materials that represent scientific research. The analysis also excluded working papers and preprints, which typically have not yet been peer reviewed

**Analysis**\
This work studied global trends to observe the growth of conference proceedings in the global scholarly literature over the past 8 years using Scopus.

The analysis suggests that scholars might have a preference for publishing conference papers. A possible reason being that they are quicker and easier to publish than journal articles or book chapters. 

The data reveals a steep increase in number of publications of conference proceedings from the start period. There was a 10 point decrease in the growth rate of publications from 2017 to 2018 followed by a 10 point increase from 2018 to 2019.

This change in growth rate did not persist however as the growth rate dropped from 2019 to 2020 and remained constant over the next 3 years. It grew by 20 points from 2022 to 2023 but then has declined over the next year.

Although its growth rate hasn't continued to increase, the total number of publications has steadily grown from ~ 10,000 at the start of the period to a peak of roughly 50,000 publications in 2023



Trends in collaboration also reveal that attitudes towards multi-collaborative works have only slightly changed as the global average number of authors per publication has increased from 3 to 4




Similarly, single-author publications have remained the majority of publications observed within the period studied. Multi-author publications however over the period 2019 to 2022 decreased but returned to pre-2019 proportions after 2022.



The global trend of the productivity of authors broadly increased over the period under study. The growth in productivity did see a decrease over the  2019 to 2022 period but then picked up going into 2023.



These global trends suggest that the COVID-19 pandemic might have been an influential factor in the the seasonal variations observed, especially those observed  over the 2019 to 2022 period. 


Global trends however might conceal lots of interesting dynamics at the level of individual countries and this could be the focus of further investigation.

Considering global collaborations, Asian partnerships are observed to be the leading proportion of international collaborations, followed by North American and then European collaborations. 



The African and mixed continent collaborations contribute the lowest share of total annual publications. Each continent in the lower bands however show different fluctuations in output. All collaboration trends however did not show any noticeable shange caused by the COVID 19 pandemic. This might just be due to the scale at which these observations are being made.


These trends reveal patterns ripe for further investigation into publishing patterns. For instance, continents/countries which showed increased conference output to a greater extent than the global trend could be clustered together to investigate their deviation.

## Conclusion
The observed trend in the global increase in publication output of conference proceedings can be due to a combination of government investment in science, ministry guidelines incentivising publishing, or academics preference for conference proceedings considering the large period under investigation.

This report presents data on research publication output by region, country, and scientific field;  and international collaboration. The first section examines global trends on publication output across Computer Science fields specifically in Artificial Intelligence. The next section focuses on collaboration among researchers within and without other regions, countries, or economies through examining coauthoring and citation patterns specifically for  artificial intelligence-related publication output and collaboration.


**Limitiations**\
The changes in publication output of conference proceedings by different countries over time might be subject to the coverage expansion of the database used and might need to be quantified.

However, even with more robust coverage and filtering, bibliometric data may retain other biases, including a bias toward English-speaking regions, countries, or economies.
